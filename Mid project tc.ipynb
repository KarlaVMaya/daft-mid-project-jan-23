```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#We bring the data, we change the names of the columns to lower case and we remove the Y from each column of years
data = pd.read_csv('Environment_Temperature_change_E_All_Data_NOFLAG.csv',encoding='latin-1')
data
```

```python
#We change the names of the columns to lower case and remove the 'Y' from each column of years
data.columns=[c.lower().replace(' ', '_') for c in data.columns]
data.columns = data.columns.str.replace('y', '')
data
```

```python
#Select only the data of 'Temperature change'
data_temp = data.loc[data.element == 'Temperature change']
data_temp
```

```python
#Select the columns to use and drop the others
data_temp=data_temp.drop(columns=['area_code', 'months_code', 'element_code', 'element', 'unit'], axis=1)
data_temp
```

```python
# Change the names of 'Area' column
data_temp.rename(columns={'area': 'country_name'}, inplace=True)
data_temp.columns
```

```python
#We deal with null values
print("General information")

data_temp_info= pd.DataFrame({"type": data_temp.dtypes, 
"unique values": data_temp.nunique(), "missing values(%)": round(data_temp.isnull().sum()/data_temp.shape[0]*100, 2)}).rename_axis(axis='rows')                       
data_temp_info
```
